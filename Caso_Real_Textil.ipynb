```python
# Función para simular un ciclo de proceso
def simular_ciclo():
    hora_inicio = 8 * 60  # Hora de inicio en minutos
    hora_termino = hora_inicio

    for actividad in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']:
        duracion = tiempos_proceso[actividad]
        hora_termino += duracion

    return hora_inicio, hora_termino

# Realizar múltiples simulaciones
num_simulaciones = 1000
total_minutos = 0
for _ in range(num_simulaciones):
    hora_inicio, hora_termino = simular_ciclo()
    total_minutos += hora_termino - hora_inicio

# Calcular la eficiencia promedio
tiempo_estandar = 300.083
eficiencia = (tiempo_estandar / (total_minutos / num_simulaciones)) * 100

print(f"La eficiencia promedio de la línea de trabajo es: {eficiencia:.2f}%")
```

```python
import random

def simulacion_inventarios(demanda_media, desviacion_estandar, inventario_inicial, capacidad_produccion, semanas):
    inventario = inventario_inicial
    costos_fijos = 7668.94
    costos_variables = 48.8
    ingresos_totales = 0
    unidades_producidas = 0
    unidades_vendidas = 0

    inventario_semanal = []
    inventario_acumulado = []

    for semana in range(1, semanas + 1):
        demanda = random.normalvariate(demanda_media, desviacion_estandar)
        demanda = max(0, round(demanda))

        if demanda <= inventario:
            unidades_vendidas_semana = demanda
            ingresos_semana = demanda * 100
            inventario -= demanda
        else:
            unidades_vendidas_semana = inventario
            ingresos_semana = inventario * 100
            inventario = 0

        produccion = min(capacidad_produccion, inventario)
        inventario += produccion
        unidades_producidas += produccion

        unidades_vendidas += unidades_vendidas_semana
        ingresos_totales += ingresos_semana

        inventario_semanal.append(inventario)
        inventario_acumulado.append(sum(inventario_semanal))

        print("Semana:", semana)
        print("Unidades vendidas semana:", unidades_vendidas_semana)
        print("Ingresos semana:", ingresos_semana)
        print("Inventario semanal:", inventario_semanal[-1])
        print("Inventario acumulado:", inventario_acumulado[-1])
        print("Unidades vendidas acumuladas:", unidades_vendidas)
        print("Ingresos totales acumulados:", ingresos_totales)
        print("------------------")

    costos_totales = costos_fijos + costos_variables * unidades_producidas
    utilidad = ingresos_totales - costos_totales

    print("Resultados de la simulación:")
    print("Unidades producidas:", unidades_producidas)
    print("Unidades vendidas:", unidades_vendidas)
    print("Inventario final:", inventario)
    print("Ingresos totales:", ingresos_totales)
    print("Costos totales:", costos_totales)
    print("Utilidad:", utilidad)

# Parámetros de la simulación
demanda_media = 10002.4103 / 100
desviacion_estandar = 4068.1557 / 100
inventario_inicial = 3000
capacidad_produccion = 350
semanas = 52

# Ejecutar la simulación
simulacion_inventarios(demanda_media, desviacion_estandar, inventario_inicial, capacidad_produccion, semanas)
```
